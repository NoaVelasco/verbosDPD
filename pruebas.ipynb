Herramienta funcional para scrapear palabras del diccionario DLE.

````python
import requests
from bs4 import BeautifulSoup
from markdownify import MarkdownConverter

url = "https://dle.rae.es/morder"
headers = {'User-Agent': 'Mozilla/5.0'}


def has_class_k5_or_k6(tag):
    return tag.name == 'p' and ('k5' in tag.get('class', []) or 'k6' in tag.get('class', []))


try:
    response = requests.get(url, headers=headers)

    if response.ok:
        soup = BeautifulSoup(response.text, 'lxml')

        resultados = soup.find('article')

        if resultados is not None:
            # eliminar la palabra que se define
            title = resultados.find('header')
            title.decompose()

            # introducir '>' antes de cada línea
            p_tags = resultados.find_all('p')
            for p_tag in p_tags:
                p_tag.insert(0, '>')

            # cambiar a negrita los números de acepción
            n_acep = resultados.find_all('span', 'n_acep')
            for acep in n_acep:
                acep.name = 'strong'

            # cambiar a cursiva los ejemplos
            ejemplos = resultados.find_all('span', 'h')
            for ejemplo in ejemplos:
                ejemplo.name = 'em'

            # cambiar a cursiva ciertas abreviaturas
            abbrs = resultados.find_all('abbr', 'c')
            for abbr in abbrs:
                abbr.name = 'em'

            # cambiar a may. preposiciones de ejemplos
            preposiciones = resultados.find_all('span', 'i1')
            for prepo in preposiciones:
                # prepo.string = prepo.string.upper()
                if prepo is not None:
                    prepo.string = prepo.string.upper()

            # cambiar hiperenlaces por negrita
            h_enlaces = resultados.find_all('a', 'a')
            for enlace in h_enlaces:
                enlace.name = 'b'

            # cambiar locuciones y composiciones por negrita
            locuciones = soup.find_all(has_class_k5_or_k6)
            for locucion in locuciones:
                # locucion.insert_after('>  \n')
                # locucion.name = 'b'
                locucion.insert_before('>  \n')

            md_converter = MarkdownConverter()
            markdown_string = md_converter.convert_soup(resultados)

            # eliminar los saltos innecesarios
            markdown = markdown_string.replace('\n\n', '\n')
            markdown = markdown.replace('\n\n', '\n')

            file = f'verbos/atender_rae.md'

            with open(file, 'w', encoding='utf-8') as f:
                f.write(f'>[!DICCIONARIO]\n>**atender**')
                f.write(markdown)
    else:
        print('Mala respuesta para', url, response.status_code)
except requests.exceptions.ConnectionError as exc:
    print(exc)
````

Tengo en mente una solución, pero para comparar le pido a la IA lo siguiente:  

> Tengo una carpeta con varios archivos txt. Cada uno es una lista de palabras. Quiero que un archivo de Python haga iteraciones con ellas, pero solo quiero que use una lista cada vez que abro el programa. ¿Cómo puedo hacer para que cada vez que lo abro sea la lista siguiente a la que abrí la última vez y que no se repitan?

Respuesta:  
> Una forma de hacerlo sería mantener un registro del último archivo que se abrió en un archivo separado. Cada vez que ejecutes el programa, puedes leer el nombre del último archivo que se abrió desde este archivo de registro y luego abrir el siguiente archivo en la carpeta. Después de procesar el archivo, puedes actualizar el archivo de registro con el nombre del archivo que acabas de abrir. ¿Te gustaría que te proporcione un ejemplo de código para hacer esto?


```python
import os

folder_path = 'ruta/a/tu/carpeta'
log_file = 'last_file.txt'

# Obtener una lista de todos los archivos txt en la carpeta
txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

# Ordenar la lista alfabéticamente para asegurarse de que el orden sea consistente
txt_files.sort()

# Leer el último archivo procesado desde el archivo de registro
if os.path.exists(log_file):
    with open(log_file, 'r') as f:
        last_file = f.read().strip()
else:
    last_file = None

# Encontrar el índice del último archivo procesado en la lista de archivos txt
if last_file in txt_files:
    last_index = txt_files.index(last_file)
else:
    last_index = -1

# Obtener el siguiente archivo en la lista
next_index = (last_index + 1) % len(txt_files)
next_file = txt_files[next_index]

# Procesar el siguiente archivo
with open(os.path.join(folder_path, next_file), 'r') as f:
    # Aquí puedes hacer lo que quieras con el contenido del archivo
    content = f.read()
    print(content)

# Actualizar el archivo de registro con el nombre del archivo que acabamos de procesar
with open(log_file, 'w') as f:
    f.write(next_file)
```
> Este código lee el nombre del último archivo procesado desde un archivo de registro llamado `last_file.txt`. Luego encuentra el siguiente archivo en la carpeta especificada y lo procesa. Finalmente, actualiza el archivo de registro con el nombre del archivo que acaba de procesar. La próxima vez que ejecutes el programa, comenzará con el siguiente archivo en la lista. 



Hago lo que tenía en mente, que es crear una lista de listas. Cada vez que se ejecute, la aplicación accederá a esta lista, que le dice cuál será la siguiente lista a utilizar. Al terminar, esta lista de listas se actualiza eliminando la primera (porque es la que acabamos de utilizar).
